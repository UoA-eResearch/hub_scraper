```python
import json
import pandas as pd
import requests_cache
requests_cache.install_cache('cache', allowable_codes=[200, 404], allowable_methods=['GET', 'POST'])
import requests
from tqdm.auto import tqdm
tqdm.pandas()
pd.set_option("display.max_colwidth", 200)
```

```python
r = requests.post("https://rhubcpapi.auckland.ac.nz/cer-graphql-service", json={
    "query": open("query.gql").read()
})
with open("reshub.json", "w") as f:
    json.dump(r.json(), f)
print(r)
if "data" not in r.json():
    print(r.json())
print(r.json()["data"].keys())
print(r.json()["data"]["articleCollection"].keys())
display(pd.DataFrame(r.json()["data"]["articleCollection"]["items"]))
display(pd.DataFrame(r.json()["data"]["subHubCollection"]["items"]))
```
