```python
import requests
from tqdm.auto import tqdm
from tqdm.contrib.concurrent import thread_map

all_policies = []

# This API only supports up to 20 pages unfortunately
for page in tqdm(range(1, 21)):
  all_policies.extend(requests.post("https://nvyjqjk6dj.execute-api.ap-southeast-2.amazonaws.com/query", json={
    "query": "",
    "scope": "policies",
    "page": page
  }).json()["results"])
len(all_policies)
```

```python
import pandas as pd
# Show whole column
pd.set_option("display.max_colwidth", 200)
pd.DataFrame(all_policies).drop_duplicates()
```

```python
from requests_html import HTMLSession
from pprint import pprint
import pandas as pd

session = HTMLSession()
links = set()

def get_links_from_page(url):
  print(url)
  for link in session.get(url).html.links:
    if link.startswith("/en/about-us/about-the-university/policy-hub/"):
      link = "https://www.auckland.ac.nz" + link
      if link not in links:
        links.add(link)
        get_links_from_page(link)

get_links_from_page('https://www.auckland.ac.nz/en/about-us/about-the-university/policy-hub.html')
pprint(links)
pd.DataFrame(links)[0].sort_values().to_csv("policy_links.csv", index=False, header=False)
```
